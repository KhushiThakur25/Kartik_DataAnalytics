```python
dis_price = np.array(df['discounted_price'])
act_price = np.array(df['actual_price'])
st = dis_price[0].replace('‚Çπ','')
act_price[0]
st
```

```python
for i in range(len(dis_price)):
    st = dis_price[i].replace('‚Çπ','')
    st = st.replace(',','')
    dis_price[i] = float(st)
for i in range(len(act_price)):
    st = act_price[i].replace('‚Çπ','')
    st = st.replace(',','')
    act_price[i] = float(st)
    
```

```python
act_price[0]-dis_price[0]
df['discounted_price'] = dis_price
df['actual_price'] = act_price
type(df['discounted_price'])
type(act_price)
```

```python
df.info()
```

```python
# pd.eval("cost_price = act_price / 2",target = df)
# pd.eval("profit = ((df.discounted_price * 100) / df.cost_price)",target = df,inplace = True)


df['cost_price'] = act_price / 2
df['profit'] = act_price - (act_price/2 + dis_price)
df
```

```python

rating_ = np.array(df['rating'])
for i in range(len(rating_)):
    rating_[i] = rating_[i].replace('|','0.0')
    rating_[i] = float(rating_[i])
```

```python
df['rating'] = pd.to_numeric(df['rating'],errors = 'coerce')
```

```python
high_rated_products = df[df['rating'] > 4.5]
high_rated_products.shape
```
