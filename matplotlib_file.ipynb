```python
# data from https://allisonhorst.github.io/palmerpenguins/

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

species = ("Adelie", "Chinstrap", "Gentoo")
penguin_means = {
    'Bill Depth': (18.35, 18.43, 14.98),
    'Bill Length': (38.79, 48.83, 47.50),
    'Flipper Length': (189.95, 195.82, 217.19),
}

x = np.arange(len(species))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in penguin_means.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Length (mm)')
ax.set_title('Penguin attributes by species')
ax.set_xticks(x + width, species)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(0, 250)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.arange(1, 6)  # X-axis values
y1 = [10, 20, 30, 40, 50]  # Y-axis 1 values
y2 = [100, 90, 80, 70, 60]  # Y-axis 2 values

# Create figure and axis
fig, ax1 = plt.subplots()

# Plot data for the first y-axis
ax1.bar(x, y1, color='b', width=0.4, label='Y1 Data')  # Bar graph for Y1
ax1.set_xlabel('X-axis')  # Label for X-axis
ax1.set_ylabel('Y1-axis', color='b')  # Label for Y1-axis
ax1.tick_params(axis='y', labelcolor='b')  # Tick color for Y1-axis
ax1.legend(loc='upper left')

# Create second y-axis
ax2 = ax1.twinx()  # Create a twin y-axis
ax2.bar(x + 0.4, y2, color='r', width=0.4, label='Y2 Data')  # Bar graph for Y2
ax2.set_ylabel('Y2-axis', color='r')  # Label for Y2-axis
ax2.tick_params(axis='y', labelcolor='r')  # Tick color for Y2-axis
ax2.legend(loc='upper right')

# Show the plot
plt.title("Bar Graph with Two Y Axes")
plt.show()
```
